## Glassdoor Scraping

```python
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
```

```python
def get_job_listings(title, num_pages):
    """
    Get job listings for a given job title and number of pages.
    Returns a list of job dictionaries.
    """
    driver = webdriver.Chrome()
    jobs=[]
    for page in range(13,num_pages+1):
        url = f"https://www.glassdoor.com/Job/{title}-jobs-SRCH_KO0,14_IP{page}.htm"
        print(url)
        driver.get(url)
        time.sleep(7)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        if page > 1:
            close_button = driver.find_element(By.CLASS_NAME, "SVGInline-svg.modal_closeIcon-svg")
            close_button.click()
        else: 
            pass
        job_listings = soup.select('li[class*="react-job-listing"]')
        for job_elem in job_listings:
            job_title = job_elem.find_all('a')[-1].span.text if job_elem.find_all('a')[-1] else ''
            company = job_elem.find('a',class_='css-l2wjgv').span.text if job_elem.find('a',class_='css-l2wjgv') else ''
            rating = job_elem.find('span',class_= "css-2lqh28 e1cjmv6j1").text if job_elem.find('span',class_= "css-2lqh28 e1cjmv6j1") else ''
            date = job_elem.find('div',class_= "d-flex align-items-end pl-std css-1vfumx3").text if job_elem.find('div',class_= "d-flex align-items-end pl-std css-1vfumx3") else ''
            location = job_elem.find('span', {'data-test': 'emp-location'}).text if job_elem.find('span', {'data-test': 'emp-location'}) else ''
            salary_range = job_elem.find('span',class_='css-1xe2xww').text if job_elem.find('span',class_='css-1xe2xww') else ''   
            #salary_avg = job_elem.find('div',class_='css-1bluz6i e2u4hf13').text if job_elem.find('div',class_='css-1bluz6i e2u4hf13') else ''
            job_url = "https://www.glassdoor.com/" + job_elem.find_all('a')[-1].get('href')
            driver.get(job_url)  # navigate to job page
            time.sleep(6)
            job_html = driver.page_source
            job_soup = BeautifulSoup(job_html, 'html.parser')
            job_descript = job_soup.find('div',{'class': 'desc'}).text if job_soup.find('div',{'class': 'desc'}).text else ''
            try:
                driver.find_element(By.XPATH,"//div[@data-test='tab' and @data-tab-type='overview']").click()
            except:
                pass
            try:
                size = driver.find_element(By.XPATH, '//label[text()="Size"]/following-sibling::span').text if driver.find_element(By.XPATH, '//label[text()="Size"]/following-sibling::span') else ''
            except:
                size = ''
            try:    
                year_founded = driver.find_element(By.XPATH, '//label[text()="Founded"]/following-sibling::span').text if driver.find_element(By.XPATH, '//label[text()="Founded"]/following-sibling::span') else ''
            except:
                year_founded = ''
            try:
                type = driver.find_element(By.XPATH, '//label[text()="Type"]/following-sibling::span').text if driver.find_element(By.XPATH, '//label[text()="Type"]/following-sibling::span') else ''
            except:
                type = ''
            try:
                industry = driver.find_element(By.XPATH, '//label[text()="Industry"]/following-sibling::span').text if driver.find_element(By.XPATH, '//label[text()="Industry"]/following-sibling::span') else ''
            except:
                industry = ''
            try:
                sector = driver.find_element(By.XPATH, '//label[text()="Sector"]/following-sibling::span').text if driver.find_element(By.XPATH, '//label[text()="Sector"]/following-sibling::span') else ''
            except:
                sector = ''
            try:
                revenue = driver.find_element(By.XPATH, '//label[text()="Revenue"]/following-sibling::span').text if driver.find_element(By.XPATH, '//label[text()="Revenue"]/following-sibling::span') else ''
            except:
                revenue = ''
            #print(f"Job Title: {job_title} | Company: {company} | Rating: {rating} | Date: {date} | Location: {location} | Salary Range: {salary_range} | Salary Avg: {salary_avg} | Size: {size} | Year Founded: {year_founded} | Type: {type} | Industry: {industry} | Sector: {sector} | Revenue: {revenue} | Description: {job_descript}")
            jobs.append({
            'title': job_title,
            'company': company,
            'rating': rating,
            'date': date,
            'location': location,
            #'salary_avg': salary_avg,
            'salary_range': salary_range,
            'type': type,
            'industry': industry,
            'sector': sector,
            'revenue': revenue,
            'size': size, 
            'year_founded': year_founded,           
            'description': job_descript
            })            
    driver.quit()
    return pd.DataFrame(jobs)
```

## Indeed scraping

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time 
import pandas as pd
def scrape_job_listings(title, num_pages):
    # chrome_options = Options()
    # chrome_options.add_argument('--headless')
    #driver = webdriver.Chrome(options=chrome_options)
    driver = webdriver.Chrome()
    jobs = []    
    for page in range(88,num_pages):
        url = f"https://www.indeed.com/jobs?q={title}&fromage=14&start={page}0"
        print(url)
        driver.get(url)
        time.sleep(6)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        job_listings = soup.select('div[class*="cardOutline"]')        
        if not job_listings:
            print("No more job listings found.")
            break        
        for job_elem in job_listings:
            job_title = job_elem.find('a').span.text if job_elem.find('a').span.text else ''
            company = job_elem.find('span',class_= "companyName").text if job_elem.find('span',class_= "companyName") else ''
            rating = job_elem.find('span',class_= "ratingNumber").text if job_elem.find('span',class_= "ratingNumber") else ''
            date = job_elem.find('span',class_= "date").text if job_elem.find('span',class_= "date") else ''
            location = job_elem.find('div',class_= "companyLocation").text if job_elem.find('div',class_= "companyLocation") else ''
            job_url = "https://br.indeed.com" + job_elem.find('a')['href']
            driver.get(job_url)  # navigate to job page
            time.sleep(4)
            job_html = driver.page_source
            job_soup = BeautifulSoup(job_html, 'html.parser')
            job_descript = job_soup.find('div', class_='jobsearch-jobDescriptionText').text.replace('\n','') if job_soup.find('div', class_='jobsearch-jobDescriptionText') else ''
            jobs.append({
                'title': job_title,
                'company': company,
                'rating':rating,
                'date':date,
                'location': location,
                'description': job_descript,
                })        
    driver.quit()
    return pd.DataFrame(jobs)
```
